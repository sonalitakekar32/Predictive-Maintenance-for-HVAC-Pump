```python
features = df.drop(['timestamp', 'machine_status'], axis=1)
target = df['machine_status']
features.head(5)
```

```python
target.head(5)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df = pd.DataFrame(scaler.fit_transform(features), columns= features.columns)
scaled_df.head(10)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size = 0.20, random_state= 100)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Using Linear Regression Model
```

```python
from sklearn.linear_model import LogisticRegression
linear_model = LogisticRegression()
linear_model.fit(x_train, y_train)
```
